```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
res=requests.get('https://www.1mg.com/categories/featured/1mg-health-products/top-products-of-vitamins-supplements-555?filter=true&pageNumber=1',headers=headers).text
```

```python
soup=BeautifulSoup(res,'lxml')
fn=str(soup)
```

```python
product_name='Revital H Woman Tablet'
product_link='/otc/health-ok-tablet-otc711094'

mark1='\\",\\"image\\":\\"'

front=product_link+mark1

fidx=fn.find(front)
new_str=fn[fidx+len(front):]
# print(new_str)
s_idx=new_str.find("\\")

print("fidx"+str(fidx))
print("s_idx"+str(s_idx))

first_idx=fidx+len(front)
print(fn[first_idx:first_idx+s_idx])
```

```python
box=soup.find_all('div','col-md-3 col-sm-4 col-xs-6 style__container___1TL2R')
box
```

```python
product_name=[]
quantity=[]
avg_star=[]
rating=[]
mrp=[]
discount=[]
price=[]
product_link=[]
product_image_link=[]
fn=str(soup)
website=[]
# mark1='{"image":"'
# mark2 ='","name":"'
# regexPattern = mark1 + '(.+?)' + mark2
# str_found = re.search(regexPattern, fn).group(1)

for i in box:
        if i.find("div","style__pro-title___2QwJy") != None:
            product_name.append(i.find("div","style__pro-title___2QwJy").text)
            pro_name=i.find("div","style__pro-title___2QwJy").text
#             print(pro_name)
        else:
            product_name.append("NA")
#         if i.find("div","style__pack-size___2JQG7") !=None:
#             quantity.append(i.find("div","style__pack-size___2JQG7").text)
#         else :
#             quantity.append("NA")
        if i.find("span","CardRatingDetail__weight-700___27w9q") !=None:
            avg_star.append(i.find("span","CardRatingDetail__weight-700___27w9q").text)
        else:
            avg_star.append("NA")
        if i.find("span","CardRatingDetail__ratings-header___2yyQW")!= None:
            rating.append(i.find("span","CardRatingDetail__ratings-header___2yyQW").text)
        else :
            rating.append("NA")
        if i.find("span","style__discount-price___25Bya") != None:
            mrp.append(i.find("span","style__discount-price___25Bya").text)
        else:
#             mrp.append("NA")
            simp=i.find("div","style__price-tag___cOxYc").text
            idx=simp.find("P")
            mrp.append(simp[idx:])
        if i.find("span","style__off-badge___2JaF-") !=None:
            discount.append(i.find("span","style__off-badge___2JaF-").text)
        else :
            discount.append("NA")
        if i.find("div","style__price-tag___cOxYc") !=None:
            simp=i.find("div","style__price-tag___cOxYc").text
            if(simp.find("P")==-1):
                price.append(i.find("div","style__price-tag___cOxYc").text)
            else:
                idx=simp.find("P")
                price.append(simp[idx+1:])
        else:
            price.append("NA")
        if i.find("a","style__product-link___UB_67")!= None:
            product_link.append("https://www.1mg.com"+i.find("a","style__product-link___UB_67").get('href'))
            pro_link=i.find("a","style__product-link___UB_67").get('href')
        else :
            product_link.append("NA")
        
        website.append("1mg")
        
        mark1='\\",\\"image\\":\\"'
        front=pro_link+mark1

        fidx=fn.find(front)
        new_str=fn[fidx+len(front):]
        s_idx=new_str.find("\\")

        first_idx=fidx+len(front)
        product_image_link.append(fn[first_idx:first_idx+s_idx])
        print(fn[first_idx:first_idx+s_idx])

        
print(len(product_image_link))
# print(product_image_link)
```

```python
df=pd.DataFrame({"Product Name":product_name,"Price":price,"Discount":discount,"MRP": mrp,"Rating":rating,"Average Star":avg_star,"Product_Link":product_link,"Website":website,"Product Image Link":product_image_link})
```
